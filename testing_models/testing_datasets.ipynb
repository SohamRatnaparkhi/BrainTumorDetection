```python
import cv2
from  keras.models import load_model
from PIL import Image
import numpy as np
import os
import time
```

```python
model = load_model("D:\College\Capstone Project 2\data\Trained_models\Big_Brain_tumor_model_1.h5")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
def plot_image(path):
    img = mpimg.imread(path)
    imgplot = plt.imshow(img)
    plt.show()
```

# Checking accuracy on other datasets on which this model is not trained but its result is known through verified sources

## SET 1

> Checking for no tumor

```python
path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\set_1\no"

incorrect_no_tumor = 0   
for x in os.listdir(path):
    try:
        path = path + "\\" + x 
        image = Image.open(path)

        image = image.resize((224,224))
        image = image.convert('RGB')
        image.save(path)

        image = cv2.imread(path)


        img = np.array(image)
        img = np.expand_dims(img, axis = 0)

        result = model.predict(img)
         
        if result.squeeze() > 0.5:
            print(f"Incorrect result on {x}\nExpected: no-tumor\nResult: Tumor")
            incorrect_no_tumor += 1
            plot_image(path)
            
        path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\set_1\no"
    except Exception as e:
        print(f"{e}  in  {x}")
        plot_image(path)
        path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\set_1\no"

print(f"Total incorrect: {incorrect_no_tumor} out of {len(os.listdir(path))} images")
```

> Checking accuracy for images with brain tumor.

```python
path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\set_1\yes"
# path = r"C:\Users\Shree\Downloads\t11.JPG"
incorrect_tumor = 0   
for x in os.listdir(path):
    try:
        path = path + "\\" + x 
        image = Image.open(path)

        image = image.resize((224,224))
        image = image.convert('RGB')
        image.save(path)

        image = cv2.imread(path)


        img = np.array(image)
        img = np.expand_dims(img, axis = 0)

        result = model.predict(img)
         
        if result.squeeze() <= 0.5:
            print(f"Incorrect result on {x}\nExpected: Tumor\nResult: No-tumor")
            incorrect_tumor += 1
            plot_image(path)
            
        path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\set_1\yes"
    except Exception as e:
        print(f"{e}  in  {x}")
        path = r"D:\College\Capstone Project 2\data\brain_tumor_dataset_testing\set_1\yes"
        plot_image(path)

print(f"Total incorrect: {incorrect_tumor} in {len(os.listdir(path))} images")
```

Result

```python
incorrect_ans = {"No-Tumor": incorrect_no_tumor,
                "Tumor": incorrect_tumor,}
# 4 incorect in 253 (97 inc + 156 corr) images

accuracy = 100 * (1 - (incorrect_no_tumor+incorrect_tumor)//(154 + 97))
print(incorrect_ans)
print(f"Accuracy of model based on this dataset is around {accuracy}%")
```
